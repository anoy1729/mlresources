```
import string 
  
file = open("pg3807.txt", "r") 
  
dictionary = dict() 
  
for line in file: 

    line = line.strip() 

    line = line.lower() 
  
    line = line.translate(line.maketrans("", "", string.punctuation)) 
  
    words = line.split(" ") 
  
    for word in words: 

        if word in dictionary: 
           
            dictionary[word] = dictionary[word] + 1
        else: 
             
            dictionary[word] = 1
  
for key in list(sorted(dictionary.keys())): 
    print(key, ":", dictionary[key]) 
```
